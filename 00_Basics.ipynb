# Manim의 구성 요소

Manim은 크게 세 가지로 구성되어 있다고 할 수 있어요.

1. Mathematical object (or Mobject)
    
    동그라미, 화살표, 좌표축, 그래프, … 화면에 나타나는 모든 객체를 의미해요.
    
2. Animation
    
    두 Mobject 사이를 보간하는 절차…라고 하면 너무 어려우니까, 일단은 그냥 흔히 우리가 생각하는 애니메이션이라고 받아들입시다. 움직이고, 모양을 바꾸고, 나타내거나 사라지는 것 모두를 의미해요.
    
3. Scene

    장면의 단위라고 생각하면 됩니다. 화면에 표시되는 모든 Mobject와 Animation은 이 Scene 안에 포함되어야 해요.

### Mobject 만들기

그럼 일단 Mobject에 해당하는 정사각형을 하나 만들어 봅시다.

````python
%%manim -qm MakeSquare

class MakeSquare(Scene):
    def construct(self):

        sq = Square(
            side_length = 5, stroke_color=GOLD_B, fill_color=GOLD_A, fill_opacity=0.5
            )
        
        self.add(sq)
        self.wait(5)
````

코드를 당장 이해하지 못해도 괜찮으니, 일단 따라 써 보세요. <br>
`Sqare()`를 통해 정사각형을 만들고, `self.add()`를 이용해 Mobject를 나타낼 수 있어요.

`self.remove()`를 사용하면 화면상의 Mobject를 없앨 수도 있답니다.

````python
%%manim -qm ShowAndHideSquare

class ShowAndHideSquare(Scene):
    def construct(self):

        sq = Square(
            side_length = 5, stroke_color=BLUE_B, fill_color=BLUE_A, fill_opacity=0.5
            )
        
        self.wait(2)
        self.add(sq)
        self.wait(5)
        self.remove(sq)
        self.wait(2)
````

### Animation 적용하기

그럼 이번에는 만든 Mobject에 Animation을 적용해 볼게요. Animation은 `self.play()`를 통해 구현할 수 있답니다.

````python
%%manim -qm SomeAnimations

class SomeAnimations(Scene):
    def construct(self):
        square = Square()

        # some animations display mobjects, ...
        self.play(FadeIn(square))

        # ... some move or rotate mobjects around...
        self.play(Rotate(square, PI/4))

        # some animations remove mobjects from the screen
        self.play(FadeOut(square))

        self.wait(1)
````

`run_time`을 사용하면 Animation이 진행되는 시간을 조정할 수 있어요.

````python
%%manim -qm RunTime

class RunTime(Scene):
    def construct(self):
        square = Square()
        self.add(square)
        self.play(square.animate.shift(UP), run_time=3)
        self.wait(1)
````

# 직접 해보기 - 도형 그리기

다양한 설정의 도형들을 그려봅시다. 그리고 Animation을 적용시킨 하나의 Scene을 만들어 보세요.

````python
%%manim -qm Shapes

class Shapes(Scene):
    def construct(self):
        circle = Circle(stroke_color='#87C2A5', fill_color='#87C2A5', fill_opacity=1)
        square = Square(stroke_color='#525893', fill_color='#525893', fill_opacity=1)
        triangle = Triangle(stroke_color='#E07A5F', fill_color='#E07A5F', fill_opacity=1)

        circle.shift(LEFT)
        square.shift(UP)
        triangle.shift(RIGHT)
        
        self.add(triangle, square, circle)
        self.play(Rotate(square, 2*PI), RunTime=2)
        self.wait(1)
````
